**Pandas - Groupby,Sorting Aggregating the data**

```python
import pandas as pd
df = pd.read_csv("/content/sample_data/california_housing_test.csv")
df.info()
```

```python
group_by_longitude = df.groupby('longitude')
group_by_longitude.head(10)
```

**Aggregating the data**

```python
longitude_means = group_by_longitude['longitude'].mean()
longitude_means
```
