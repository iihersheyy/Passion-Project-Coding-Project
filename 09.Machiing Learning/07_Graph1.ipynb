```python
import pandas as pd
import numpy
from matplotlib import pylab
import seaborn

df = seaborn.load_dataset("mpg")
df.head(3)
```

**Plot Group**

```python
df['origin'].value_counts().plot(kind="bar")
```

**Line Graph**

```python
df['origin'].value_counts().plot(kind="line")
```

*Graph : barh*

```python
df['origin'].value_counts().plot(kind="barh")
```

**Graph Type : Pie**

```python
df['origin'].value_counts().plot(kind="pie")
```

**Graph Type : hist**

```python
df['origin'].value_counts().plot(kind="hist")
```

**Graph Type : Box **

```python
df['origin'].value_counts().plot(kind="box")
```

**Graph Type : Scatter Plot**

```python
df.plot(kind="scatter", x="mpg", y="model_year")
```

**Graph Type : Area**

```python
df['model_year'].plot(kind="area")
```

**Graph Type : hexbin**

```python
df.plot(kind="hexbin", x="mpg", y="model_year", gridsize=20)
```
